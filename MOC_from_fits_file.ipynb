```python
from mocpy import MOC
from astropy.io import fits
import numpy as np
```

```python

image_path = 'notebooks/demo-data/FITS_image_16_with_BLANK.fits'
with fits.open(image_path) as hdulist:
    print(hdulist.info())
    moc_from_file = MOC.from_image(header=fits.getheader(image_path),
                                   hdu=hdulist[0].data,
                                   moc_order=17,
                                   mask_arr=None)
```

```python
moc_from_file.write('/home/baumann/mocpy/blank_test.fits', format='fits', write_to_file=True)
```

```python
moc_from_file.plot()
```
