```python
import warnings 
warnings.filterwarnings('ignore')
from mocpy import MOC
```

```python
from astroquery.vizier import Vizier

viz = Vizier(columns=['*', '_RAJ2000', '_DEJ2000'])
viz.ROW_LIMIT = -1
table = viz.get_catalogs('I/293/npm2cros')[0]
```

```python
moc = MOC.from_file('demo-data/P-GALEXGR6-AIS-FUV.fits')
moc.plot()
```

```python
print(table)
```

```python
filtered_table = moc.filter_table(table, '_RAJ2000', '_DEJ2000')
```

```python
print(filtered_table)
```

```python
m1 = MOC.from_table(table, '_RAJ2000', '_DEJ2000', 6)

m2 = MOC.from_table(filtered_table, '_RAJ2000', '_DEJ2000', 6)

m1.plot()
m2.plot()
```
