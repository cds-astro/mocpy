# Query the sources of a VizieR table that are in a sky region

```python
import astropy.units as u
from mocpy import MOC
```

First, we retrieve the MOC for the VizieR table `VIII/84/7c` -- see [2007MNRAS.382.1639H](https://ui.adsabs.harvard.edu/abs/2007MNRAS.382.1639H).

```python
moc_radio_sources = MOC.from_vizier_table("VIII/84/7c", max_depth=7)
```

```python
moc_radio_sources.display_preview()
```

Then, we query the Hipparcos catalogue to get all its sources in the MOC coverage of the radio survey `VIII/84/7c`.

```python
hipparcos_in_7C_coverage = moc_radio_sources.query_vizier_table(
    "I/239/hip_main", max_rows=100000
)
hipparcos_in_7C_coverage
```

We could also restrict the columns with the `select` argument. Finally, we plot the positions of the sources in the table to check if they lie in the MOC coverage

```python
moc_hipparcos_in_7C_coverage = MOC.from_lonlat(
    hipparcos_in_7C_coverage["_RAJ2000"].T * u.deg,
    hipparcos_in_7C_coverage["_DEJ2000"].T * u.deg,
    max_norder=7,
)
moc_hipparcos_in_7C_coverage.display_preview()
```
